```python
import numpy as np
import cv2
import cmath
import math

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
from skimage.transform import resize

def ConvertToGaryscale(ColoredImage):
    
    Red, Green, Blue = ColoredImage[:,:,0], ColoredImage[:,:,1], ColoredImage[:,:,2]
    GrayscaleImage = 0.299 * Red + 0.587 * Green + 0.114 * Blue

    return GrayscaleImage

einstein = imread("Downloads/IMG-20220313-WA0173.jpg")
einstein = ConvertToGaryscale(einstein)


marlyin = imread("Downloads/b59b282280e8826c2c421c291f33548b.jpg")
marlyin = ConvertToGaryscale(marlyin)

einstein = resize(einstein, marlyin.shape,anti_aliasing = True)

print(einstein.shape)
print(marlyin.shape)

# img = cv2.imread("Downloads/IMG-20220313-WA0173.jpg", cv2.IMREAD_GRAYSCALE)

def fft(img):
    rows = len(img)
    if rows == 1:
        return img

    Feven, Fodd = fft(img[0::2]), fft(img[1::2])
    combined = [0] * rows
    for i in range(int(rows/2)):
        expTerm = cmath.exp((-2.0 * cmath.pi * 1j * i) / rows)
        combined[i] = Feven[i] + (expTerm * Fodd[i])
        combined[i + int(rows/2)] = Feven[i] - expTerm * Fodd[i]
    return combined

def pad2(img):
    rows, cols = np.shape(img)
    M, N = 2 ** int(math.ceil(math.log(rows, 2))), 2 ** int(math.ceil(math.log(cols, 2)))
    F = np.zeros((M,N), dtype = img.dtype)
    F[0:rows, 0:cols] = img
    return F 

def fft2(img):
    img = pad2(img)
    return np.transpose(fft(np.transpose(fft(img))))

def ifft2(img,rows,cols):
    img = fft2(np.conj(img))
    img = np.matrix(np.real(np.conj(img)))/(rows*cols)
    return img[0:rows, 0:cols]

def high_pass_filter(img):
    row, col = img.shape
    mask = np.zeros((row, col))
    hpf=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])  
    mask[:3,:3] = hpf

    mask = fft2(np.array(mask))
    mask = fftshift(mask)
  
    fourier = fft2(np.array(img))
    fourier = fftshift(fourier)

    img_new=np.abs(ifft2(fftshift(mask*fourier),rows,cols))

    return img_new


def low_pass_filter(img2):
    row, col = img2.shape
    mask2 = np.zeros((row, col))
    avgMask2= np.ones((9,9))/81
    mask2[:9,:9] = avgMask2

    mask2 = fft2(np.array(mask2))
    mask2 = fftshift(mask2)

    fourier2 = fft2(np.array(img2))
    fourier2 = fftshift(fourier2)

    img_new2=np.abs(ifft2(fftshift(mask2*fourier2),rows,cols))
    return img_new2


def fftshift(img):
    rows, cols = img.shape
    firstQuad, secondQuad = img[0: int(rows/2), 0: int(cols/2)], img[int(rows/2): rows, 0: int(cols/2)]
    thirdQuad, fourthQuad = img[0: int(rows/2), int(cols/2): cols], img[int(rows/2): rows, int(cols/2): cols]
    shiftedImg = np.zeros(img.shape,dtype = img.dtype)
    shiftedImg[int(rows/2): rows, int(cols/2): cols], shiftedImg[0: int(rows/2), 0: int(cols/2)] = firstQuad, fourthQuad
    shiftedImg[int(rows/2): rows, 0: int(cols/2)], shiftedImg[0: int(rows/2), int(cols/2): cols]= thirdQuad, secondQuad
    return shiftedImg

rows,cols = np.shape(einstein)


ein_high = high_pass_filter(einstein)


marlyin_low = low_pass_filter(marlyin)

# cv2.imwrite("Downloads/lowpass_1.jpg",lowPassImage)
# cv2.imwrite("Downloads/highpass_1.jpg", highPassImage)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2,figsize = (16, 16))

axs[0].imshow(marlyin,cmap = "gray")
axs[1].imshow(marlyin_low,cmap = "gray")

```
