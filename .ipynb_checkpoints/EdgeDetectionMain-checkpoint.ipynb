```python
from ConvertToGrayscale import ConvertToGaryscale
from EdgeDetection import Canny,Prewitt,Sobel,Roberts
import matplotlib.pyplot as plt

```

```python
image=plt.imread('apple.jpeg')
```

```python
SobelImage=Sobel(image)
print('Sobel Done')
RobertsImage=Roberts(image)
print('Roberts Done')
PrewittImage=Prewitt(image)
print('Prewitt Done')
CannyImage=Canny(image)
print('Canny Done')
```

```python
fig = plt.figure(figsize=(10, 7))
fig.add_subplot(2, 2, 1)
plt.imshow(SobelImage,cmap='gray')
plt.axis('off')
plt.title("Sobel")
fig.add_subplot(2, 2, 2) 
plt.imshow(RobertsImage,cmap='gray')
plt.axis('off')
plt.title("Roberts")
fig.add_subplot(2, 2, 3) 
plt.imshow(PrewittImage,cmap='gray')
plt.axis('off')
plt.title("Prewitt")
fig.add_subplot(2, 2, 4) 
plt.imshow(CannyImage,cmap='gray')
plt.axis('off')
plt.title("Canny")
plt.show()
```
